#2 - Qual foi o ticket médio do mercado em Junho de 2020?

```python
# Remove caracteres não numéricos da coluna "Ticket Médio" e converte para tipo numérico
arquivo['Ticket Médio'] = arquivo['Ticket Médio'].apply(lambda x: float(re.sub(r'[^\d.]', '', x)))


# Converte a coluna "Ano" para string
arquivo['Ano'] = arquivo['Ano'].astype(str)

# Seleciona as linhas em que a coluna "Ano" começa com "2020"
df_2020 = arquivo.loc[arquivo['Ano'].str.startswith('2020')]


# Separa o mês de Junho
df_junho_2020 = df_2020.loc[arquivo["Mês"].str.startswith("Jun")]

# Separa apenas a coluna do "Ticket Médio"
ticket_medio = df_junho_2020['Ticket Médio']

print("Ticket Médio do mês de Junho de 2020:",format(ticket_medio))
```

#3 - Qual foi o mês com o maior número de clientes no ano de 2021?

```python
# Converte a coluna "Ano" para string
arquivo['Ano'] = arquivo['Ano'].astype(str)

# Seleciona as linhas em que a coluna "Ano" começa com "2021"
df_2021 = arquivo.loc[arquivo['Ano'].str.startswith('2021')]

# Encontra o índice do valor máximo na coluna "Número de Clientes"
indice_maximo = df_2021["Número de Clientes"].idxmax()

# Obtém o valor do mês com o maior número de clientes
mes_maximo = df_2021.loc[indice_maximo, "Mês"]

print("Mês com o maior número de clientes:", str(mes_maximo))
```
